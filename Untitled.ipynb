    <h2 style="color:#2c3e50;">🔹Handling Data Inconsistency</h2>
<p>
The medical_conditions column contains inconsistent categorical values, including medical condition names, variations of "No," and unknown or unclear responses. To ensure uniformity, all text values are standardized by converting them to lowercase and removing leading and trailing whitespace.

A predefined list of non-affected responses (e.g., "no", "na","nothing","none", "never", "না", "না।" etc ) is used to classify entries as "No", indicating absence of any medical condition or side effect. Entries that are missing, unknown, or cannot be clearly interpreted are labeled as "Unknown" to preserve data integrity.

All remaining values, including disease names and descriptive responses, are categorized as "Yes", indicating the presence of a reported medical condition or side effect.

Steps for handling data inconsistency are as follows:</p>
<ul>
1. Convert all text to lowercase</ul>
<ul>2. Strip leading and trailing whitespace</ul>
<ul>3. Define a list of "No" and "Unknown" values</ul>
<ul>4. Classify values into Yes/No</ul>

```python
#1. Convert all text to lowercase

df["medical_conditions"] = df["medical_conditions"].str.lower()
```

```python
# 2. Strip leading and trailing whitespace

df["medical_conditions"] = df["medical_conditions"].str.strip()
```

```python
# 3. Define a list of "No" values

no_values = ["no","na","negative","never","no condition","no conditions","no disease","no medical condition","no.","none","nope","not yet by the grace of Allah","nothing","nothing seriuos","n0","না","নাহ"]
```

```python
# Define a list of "Unknown" value

unknown_values = ["unknown"]
```

```python
# 4. Classify values into Yes/No/Unknown
def categorize(x):
    if x in no_values:
        return "No"
    elif x in unknown_values:
        return "Unknown"
    else:
        return "Yes"
df["medical_conditions"] = df["medical_conditions"].apply(categorize)
```

<div style="background: linear-gradient(to right, #eef2f3, #dfe9f3); padding:15px; border-radius:15px; font-family:Arial;">
<p>
Missing values in the dataset are handled by applying appropriate replacement strategies based on the nature of each feature. Since the affected columns are mostly categorical or text-based, meaningful default categories are assigned to preserve data consistency and avoid bias.</p>
<ul>    
<li>Gender: Missing values are replaced with "Prefer not to say" to respect non-disclosure and maintain an existing valid category structure.
<li>Occupation: Missing entries are filled with "Unknown" to indicate unavailable information without assuming a specific job type.
<li>sleep_latency: Missing values are replaced with "Unknown" since the response is categorical (time ranges) and cannot be accurately inferred.
<li>sleep_reasons: Missing values are filled with "Not specified" to clearly indicate that no reason was provided.
<li>sleep_sideeffects: Missing values are replaced with "No side effects reported" to distinguish between absence of response and absence of symptoms.
<li>coping_strategies: Missing values are filled with "No coping strategy reported" to reflect non-response while maintaining interpretability.
</li>
</ul>
</div>

```python
# Gender (male, female, prefer not to say)

df["gender"] = df["gender"].fillna("Prefer not to say")
```

```python
# Occupation

df["occupation"] = df["occupation"].fillna("Unknown")
```
