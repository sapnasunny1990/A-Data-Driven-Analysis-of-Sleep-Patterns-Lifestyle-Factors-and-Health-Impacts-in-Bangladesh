<div  font-family:Arial;>
    <h2 style="color:#2c3e50;">🔹Handling Data Inconsistency</h2>
<p>
The medical_conditions column contains inconsistent categorical values, including medical condition names, variations of "No," and unknown or unclear responses. To ensure uniformity, all text values are standardized by converting them to lowercase and removing leading and trailing whitespace.

A predefined list of non-affected responses (e.g., "no", "na","nothing","none", "never", "না", "না।" etc ) is used to classify entries as "No", indicating absence of any medical condition or side effect. Entries that are missing, unknown, or cannot be clearly interpreted are labeled as "Unknown" to preserve data integrity.

All remaining values, including disease names and descriptive responses, are categorized as "Yes", indicating the presence of a reported medical condition or side effect.

Steps for handling data inconsistency are as follows:</p>
<ul>
1. Convert all text to lowercase</ul>
<ul>2. Strip leading and trailing whitespace</ul>
<ul>3. Define a list of "No" and "Unknown" values</ul>
<ul>4. Classify values into Yes/No</ul>

```python
#1. Convert all text to lowercase

df["medical_conditions"] = df["medical_conditions"].str.lower()
```

```python
# 2. Strip leading and trailing whitespace

df["medical_conditions"] = df["medical_conditions"].str.strip()
```

```python
# 3. Define a list of "No" values

no_values = ["no","na","negative","never","no condition","no conditions","no disease","no medical condition","no.","none","nope","not yet by the grace of Allah","nothing","nothing seriuos","n0","না","নাহ"]
```

```python
# Define a list of "Unknown" value
unknown_values = ["unknown"]
```

```python
# 4. Classify values into Yes/No/Unknown
def categorize(x):
    if x in no_values:
        return "No"
    elif x in unknown_values:
        return "Unknown"
    else:
        return "Yes"
df["medical_conditions"] = df["medical_conditions"].apply(categorize)
```

<div  font-family:Arial;>
    <h2 style="color:#2c3e50;font size=10">1.BMI Column</h2>
<p>
Body Mass Index (BMI) was calculated to assess the relationship between an individual's weight and height. Since the given height was in feet (5.1 ft), it was first converted into meters using the standard conversion factor (1 foot = 0.3048 meters). The converted height (1.554 meters) was then squared as required by the BMI formula.
<ul>
    <b><I>BMI = Weight/Height</I></b></ul></div>
</p>   

```python
df["bmi"] = round(df["weight"] / (df["height"]*0.3048)**2,2)
```

```python
print(df["bmi"])
```

<div  font-family:Arial;>
    <h2 style="color:#2c3e50;font size=10">2.BMI Category Column</h2>
<p>
The BMI Category column is a derived categorical feature created from the calculated Body Mass Index (BMI) values. It is used to group individuals into standard weight status categories based on their BMI score, making the data easier to interpret and analyze.
The categories are typically defined as:</p>
<ul>
<li>Underweight: BMI < 18.5</li>
<li>Normal weight: 18.5 ≤ BMI < 24.9</li>
<li>Overweight: 25 ≤ BMI < 29.9</li>
<li>Obese: BMI ≥ 30</li>
</ul>
<p>
This transformation converts a continuous numerical variable (BMI) into a categorical variable, which is useful for exploratory data analysis (EDA), visualization, and identifying health risk patterns across different population groups. It helps in comparing distributions and understanding how many individuals fall into each weight category.
</p>

```python
df["bmi_category"] = df["bmi"].apply(
    lambda x: "Underweight" if x < 18.5
    else "Normal" if x < 25
    else "Overweight" if x < 30
    else "Obese"
)
```

```python
print(df["bmi_category"])
```

<div  font-family:Arial;>
    <h2 style="color:#2c3e50;">🔹Feature Transformation</h2>
<p>
Bedtime, wake-up time, sleep latency, and sleep duration variables were transformed from categorical time-based responses into numerical values using predefined mapping techniques. Since these variables were originally recorded in ranges (e.g., "10 PM – 12 AM", "6–8 hours") and categorical time intervals (e.g., "Less than 15 minutes", "After 2 AM"), each category was assigned a representative numeric value based on its midpoint or logical approximation. This transformation ensured that all time-related and duration-based features were represented in a consistent numeric format</p>
<p>
The columns sleep_reasons, sleep_sideeffects, and coping_strategies contain multiple responses in a single cell, separated by a delimiter(;). To analyze these multi-choice responses effectively, feature transformation using explode() is applied.
</p>

```python
# Mapping bedtime

bedtime_mapping = {
    "Before 10 PM": 21,      # 9 PM (approx)
    "10 PM - 12 AM": 23,     # average of 22–24 → 23
    "12 AM - 2 AM": 25,       # average of 24–26 → 25
    "After 2 AM": 27          # representative value
}

df_new["bedtime_hour"] = df_new["bedtime"].map(bedtime_mapping)
```

```python
df_new["bedtime_hour"] 
```

```python
# Mapping wake_workdays

wake_mapping  = {
    "Before 6 AM": 5,       # 5 AM (approx)
    "6 AM - 8 AM": 7,       # average of 6–8 → 7
    "8 AM - 10 AM": 9,      # average of 8–10 → 9
    "After 10 AM": 11       # representative value
}

df_new["wake_workdays_hour"] = df_new["wake_workdays"].map(wake_mapping)
```

```python
df_new["wake_workdays_hour"]
```

```python
# Mapping bedtime_weekend

bedtime_weekend_mapping = {
    "Before 10 PM": 21,      # 9 PM (approx)
    "10 PM - 12 AM": 23,     # average of 22–24 → 23
    "12 AM - 2 AM": 25,       # average of 24-26 → 25
    "After 2 AM": 27          # representative value
}

df_new["bedtime_weekend_hour"] = df_new["bedtime_weekend"].map(bedtime_weekend_mapping)
```
